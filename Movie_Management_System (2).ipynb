```python
import json
import os
from datetime import datetime

# ================================
# FILE HANDLING
# ================================
USERS_FILE = "users.json"
MOVIES_FILE = "movies.json"
BORROW_FILE = "borrowings.json"


def load_file(filename):
    if not os.path.exists(filename):
        return []
    with open(filename, "r") as file:
        try:
            return json.load(file)
        except json.JSONDecodeError:
            print(f"Error: {filename} is corrupted. Initializing with empty data.")
            return []


def save_file(filename, data):
    with open(filename, "w") as file:
        json.dump(data, file, indent=4)


# ================================
# USER MANAGEMENT
# ================================
#users might be dict, finding the next not-used id as an integer as new id

def generate_user_id(users):
    if not users:
        return 1
    return max(user["id"] for user in users) + 1

## Need to rewrite, we don't want to write in the source file.
def add_user():
    users = load_file(USERS_FILE)
    name = input("Enter user name: ").strip()

    if not name:
        print("Name cannot be empty.")
        return

    user_id = generate_user_id(users)
    users.append({"id": user_id, "name": name})

    save_file(USERS_FILE, users)
    print(f"User added successfully! User ID: {user_id}")


def view_users():
    users = load_file(USERS_FILE)

    if not users:
        print("No users found.")
        return

    print("\n--- Users List ---")
    for user in users:
        print(f"ID: {user['id']} | Name: {user['name']}")
    print("------------------\n")


# ================================
# MOVIE MANAGEMENT
# ================================
def generate_movie_id(movies):
    if not movies:
        return 1
    return max(movie["id"] for movie in movies) + 1


def add_movie():
    movies = load_file(MOVIES_FILE)

    title = input("Enter movie title: ").strip()
    director = input("Enter director: ").strip()
    year = input("Enter year of release: ").strip()
    copies = input("Enter available copies: ").strip()

    if not title or not director or not year or not copies.isdigit():
        print("Invalid input. Please try again.")
        return

    movie_id = generate_movie_id(movies)

    movies.append({
        "id": movie_id,
        "title": title,
        "director": director,
        "year": year,
        "copies": int(copies)
    })

    save_file(MOVIES_FILE, movies)
    print(f"Movie added successfully! Movie ID: {movie_id}")


def view_movies():
    movies = load_file(MOVIES_FILE)

    if not movies:
        print("No movies found.")
        return

    print("\n--- Movies List ---")
    for m in movies:
        print(f"ID: {m['id']} | Title: {m['title']} | Director: {m['director']} | "
              f"Year: {m['year']} | Copies: {m['copies']}")
    print("-------------------\n")


# ================================
# BORROWING SYSTEM
# ================================
def borrow_movies():
    users = load_file(USERS_FILE)
    movies = load_file(MOVIES_FILE)
    borrowings = load_file(BORROW_FILE)

    ## rewrite!!!!
    try:
        user_id = int(input("Enter User ID: "))
    except ValueError:
        print("Invalid User ID.")
        return

    if not any(u["id"] == user_id for u in users):
        print("User not found.")
        return

    ## rewrite!!!!

    movie_ids_raw = input("Enter Movie IDs to borrow (comma separated): ")
    try:
        movie_ids = [int(m.strip()) for m in movie_ids_raw.split(",")]
    except ValueError:
        print("Invalid movie ID format.")
        return

    unavailable = []

    # Check availability
    for mid in movie_ids:
        movie = next((m for m in movies if m["id"] == mid), None)
        if not movie:
            unavailable.append(f"{mid} (not found)")
        elif movie["copies"] <= 0:
            unavailable.append(f"{mid} ({movie['title']} - no copies left)")

    if unavailable:
        print("\nCannot borrow the following movies:")
        for item in unavailable:
            print(" -", item)
        return

    # Borrowing is possible ‚Üí Update movie copies
    for mid in movie_ids:
        for m in movies:
            if m["id"] == mid:
                m["copies"] -= 1

    # Check if user already has a borrowing record
    record = next((b for b in borrowings if b["user_id"] == user_id), None)

    if record:
        record["movie_ids"].extend(movie_ids)
    else:
        borrowings.append({
            "user_id": user_id,
            "movie_ids": movie_ids,
            "date": datetime.now().strftime("%Y-%m-%d")
        })

    save_file(MOVIES_FILE, movies)
    save_file(BORROW_FILE, borrowings)
    print("Movies borrowed successfully!")


def return_movies():
    movies = load_file(MOVIES_FILE)
    borrowings = load_file(BORROW_FILE)

    try:
        user_id = int(input("Enter User ID: "))
    except ValueError:
        print("Invalid User ID.")
        return

    record = next((b for b in borrowings if b["user_id"] == user_id), None)
    if not record:
        print("This user has no borrowed movies.")
        return

    movie_ids_raw = input("Enter Movie IDs to return (comma separated): ")
    try:
        movie_ids = [int(m.strip()) for m in movie_ids_raw.split(",")]
    except ValueError:
        print("Invalid movie ID format.")
        return

    for mid in movie_ids:
        if mid in record["movie_ids"]:
            record["movie_ids"].remove(mid)

            # Increase copies back
            for m in movies:
                if m["id"] == mid:
                    m["copies"] += 1

    # If no movies left, remove the record
    borrowings = [b for b in borrowings if b["movie_ids"]]

    save_file(MOVIES_FILE, movies)
    save_file(BORROW_FILE, borrowings)
    print("Movies returned successfully!")


def list_borrowed_movies():
    movies = load_file(MOVIES_FILE)
    borrowings = load_file(BORROW_FILE)

    try:
        user_id = int(input("Enter User ID: "))
    except ValueError:
        print("Invalid User ID.")
        return

    record = next((b for b in borrowings if b["user_id"] == user_id), None)
    if not record:
        print("This user has not borrowed any movies.")
        return

    print("\n--- Borrowed Movies ---")
    print(f"Borrowing Date: {record['date']}")

    for mid in record["movie_ids"]:
        movie = next((m for m in movies if m["id"] == mid), None)
        if movie:
            print(f"ID: {mid} | Title: {movie['title']}")

    print("-----------------------\n")


# ================================
# ERROR PROTOCOL (These are notes, not executable code)
# ================================

# missing files (This logic is now handled inside load_file)
# if not os.path.exists(filename):
#     return []

# corrupted file (This logic is now handled inside load_file)
# try:
#     return json.load(file)
# except json.JSONDecodeError:
#     print(f"Error: {filename} is corrupted. Replacing with empty file.")
#     return []

# user input error (Example of a check within a function)
# if not name:
#     print("Name cannot be empty.")
#     return

# wrong input (Example of a check within a function)
# try:
#     user_id = int(input("Enter User ID: "))
# except ValueError:
#     print("Invalid User ID. Please enter a number.")
#     return

# movie not found error (Example of a check within a function)
# if not movie:
#     unavailable.append(f"{mid} (not found)")






# ================================
# MAIN MENU
# ================================
def main_menu():
    while True:
        print("\n===== MOVIE MANAGEMENT SYSTEM =====")
        print("1. Add User")
        print("2. View Users")
        print("3. Add Movie")
        print("4. View Movies")
        print("5. Borrow Movies")
        print("6. Return Movies")
        print("7. List Borrowed Movies for a User")
        print("8. Exit")
        choice = input("Select an option: ")

        if choice == "1":
            add_user()
        elif choice == "2":
            view_users()
        elif choice == "3":
            add_movie()
        elif choice == "4":
            view_movies()
        elif choice == "5":
            borrow_movies()
        elif choice == "6":
            return_movies()
        elif choice == "7":
            list_borrowed_movies()
        elif choice == "8":
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")




# Run the program
main_menu()
```
